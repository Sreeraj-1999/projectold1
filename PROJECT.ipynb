```python
#IMPORT DATASET

import pandas as pd
df=pd.read_csv('/content/b_depressed.csv')
df
```

```python
#CHECK FOR MISSING VALUES
df.isna().sum()
```

```python
df.dtypes
```

```python
# MISSING VALUE HANDLING
df['no_lasting_investmen']=df['no_lasting_investmen'].fillna(df['no_lasting_investmen'].mode()[0])
```

```python
#CHECK WHETHER IMBALANCED DATA EXISTS
df['depressed'].value_counts()
```

```python
X=df.iloc[:,2:-1]
y=df.iloc[:,-1]
```

```python
#SCALING
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
X=mm.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#KNEIGHBORS ALGORITHM BEFORE HYPERPARAMETER TUNING
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#KNEIGHBORS ALGORITHM BEFORE HYPERPARAMETER TUNING
from sklearn.model_selection import GridSearchCV
knn1=KNeighborsClassifier()
params={'n_neighbors':[3,5,7,9],'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute']}
clf=GridSearchCV(knn1,params,cv=5,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
knn=KNeighborsClassifier(algorithm= 'auto', n_neighbors= 9, weights= 'uniform')
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#SVM BEFORE HYPERPARAMETER TUNING
from sklearn.svm import SVC
sv=SVC(kernel='poly')
sv.fit(X_train,y_train)
y_pred1=sv.predict(X_test)
print(classification_report(y_test,y_pred1))
```

```python
#SVM AFTER HYPERPARAMETER TUNING (refit=True,verbose=2)
sv=SVC()
params={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf','poly','sigmoid']}
clf=GridSearchCV(sv,params,cv=5,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
from sklearn.svm import SVC
sv=SVC(C=0.1,kernel='rbf',gamma=1)
sv.fit(X_train,y_train)
y_pred1=sv.predict(X_test)
print(classification_report(y_test,y_pred1))
```

```python

from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred2=nb.predict(X_test)
print(classification_report(y_test,y_pred2))
```

```python
#RANDOM FOREST BEFORE HYPERPARAMETER TUNING
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=65,random_state=1)
rf.fit(X_train,y_train)
y_pred3=nb.predict(X_test)
print(classification_report(y_test,y_pred3))
```

```python
rf=RandomForestClassifier()
params1={'n_estimators':[25,50,100,150],'max_features':['sqrt','auto'],'max_depth':[2,4],'min_samples_split':[2,5],'min_samples_leaf':[1,2]}
clf=GridSearchCV(rf,params1,cv=5,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
#RANDOM FOREST AFTER HYPERPARAMETER TUNING
from math import sqrt
rf=RandomForestClassifier(n_estimators=25,max_depth=4,max_features='sqrt',min_samples_leaf=1,min_samples_split=5,random_state=1)
rf.fit(X_train,y_train)
y_pred3=nb.predict(X_test)
print(classification_report(y_test,y_pred3))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_depth=3,random_state=1)
clf.fit(X_train,y_train)
y_pred4=clf.predict(X_test)
print(classification_report(y_test,y_pred4))
```

```python
clf=DecisionTreeClassifier()
params2={'criterion':['gini', 'entropy', 'log_loss'],'splitter':['best','random'],'max_depth':[2,3,4,5,None],'min_samples_split':[2,4,6],'min_samples_leaf':[1,2,3,4],'min_weight_fraction_leaf':[0.0,1.0,2.0,3.0],'max_features':['auto','sqrt','log2'],'max_leaf_nodes':[None,1,2,3,4]}
clf=GridSearchCV(clf,params2,cv=5,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_features='auto',max_leaf_nodes=3,max_depth=3,random_state=1,splitter='best',min_weight_fraction_leaf=0.0,min_samples_split=4,min_samples_leaf=1)
clf.fit(X_train,y_train)
y_pred4=clf.predict(X_test)
print(classification_report(y_test,y_pred4))
```
